```python
import pandas as pd
import plotly.express as px
```

```python
unemp = pd.read_excel("Unemployment_by_ACLED_regions_v2.xlsx", sheet_name=None)
riots = pd.read_excel("number_of_political_violence_events_by_country-year_as-of-19Sep2025.xlsx")
```

```python
keys = list(unemp.keys())
```

```python
unemp = pd.read_excel("Unemployment_by_ACLED_regions_v2.xlsx", sheet_name=keys)
```

```python
target_columns = ['Country Name', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']
```

```python
unemp['Africa'][target_columns]
```

```python
Africa = unemp['Africa'][target_columns]
Africa = Africa.fillna(0)
Africa
```

```python
years = target_columns[1:]
```

```python
summary_list = []
for sheet, df in unemp.items():
    total_row = df[years].median().to_frame().T
    total_row.insert(0, 'Region', sheet)

    summary_list.append(total_row)

summary_df = pd.concat(summary_list, ignore_index=True)

summary_df
```

```python
region_dict = {}

for sheet_name, df in unemp.items():
    countries = df['Country Name'].tolist()
    for country in countries:
        region_dict[country] = sheet_name

region_dict
```

```python
riots_pivot = riots.pivot(index='COUNTRY', columns='YEAR', values='EVENTS')
riots_pivot.columns.name = None
riots_pivot = riots_pivot.reset_index()
riots_pivot.columns = riots_pivot.columns.map(str)
riots_pivot
```

```python
riots_columns = ['COUNTRY'] + years
riots_columns
```

```python
riots_pivot = riots_pivot[riots_columns]
riots_pivot = riots_pivot.set_index(riots_pivot.columns[0])
riots_pivot
```

```python
riots_region = pd.DataFrame(index=['Africa',
                                'Asia-Pacific',
                                'Europe_and_Central_Asia',
                                'Latin_America_and_Caribbean',
                                'Middle_East_and_North_Africa',
                                'US_and_Canada'],
                           columns=['2008',
                                    '2009',
                                    '2010',
                                    '2011',
                                    '2012',
                                    '2013',
                                    '2014',
                                    '2015',
                                    '2016',
                                    '2017',
                                    '2018',
                                    '2019',
                                    '2020',
                                    '2021',
                                    '2022',
                                    '2023'])

riots_region
```

```python
for country in region_dict.keys():
    for col in riots_region.columns:
        if country in riots_pivot.index:
            if pd.isna(riots_region.loc[region_dict[country], col]):
                if pd.notna(riots_pivot.loc[country, col]):
                    riots_region.loc[region_dict[country], col] = riots_pivot.loc[country, col]
            else:
                if pd.notna(riots_pivot.loc[country, col]):
                    riots_region.loc[region_dict[country], col] += riots_pivot.loc[country, col]

riots_region
```

```python
riots_region = riots_region.fillna(0)
riots_region
```

```python
summary_df.to_csv("unemployment_rate_data.csv", index=False)
```

```python
riots_region = riots_region.reset_index(names="Region")
riots_region.to_csv("riots_data.csv", index=False)
```
