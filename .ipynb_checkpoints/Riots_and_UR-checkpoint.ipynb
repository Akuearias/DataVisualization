```python
import pandas as pd
```

```python
ur = pd.read_csv('Unemployment_Rate.csv')
ur
```

```python
riots = pd.read_csv('Riots.csv')
riots
```

```python
df = ur.T
df.columns = df.iloc[0]
df = df.drop(df.index[0])
df
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from distinctipy import get_colors

fig, ax = plt.subplots(figsize=(16, 9))
palette = get_colors(25)


for i, col in enumerate(df.columns):
    ax.plot(df.index, df[col], label=col, color=palette[i])

ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax.set_xlabel('Year')
ax.set_ylabel('Unemployment rate (%)')

fig.savefig('Unemployment_rate.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
palette = get_colors(25)

fig, ax = plt.subplots(figsize=(16, 16))
for i, (country, group) in enumerate(riots.groupby("COUNTRY")):
    ax.plot(group["YEAR"], group["VALUE"], marker="o", label=country, color=palette[i])

ax.set_xlabel("Year")
ax.set_ylabel("Value")

ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

fig.savefig('Riots.png', dpi=300, bbox_inches='tight')
plt.show()
```
