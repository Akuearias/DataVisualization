```python
import geopandas as gpd
```

```python
world = gpd.read_file("https://raw.githubusercontent.com/nvkelso/natural-earth-vector/master/110m_cultural/ne_110m_admin_0_countries.shp")
world.plot()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Week11_Example.csv")
data
```

```python
merged = world.merge(data, left_on="NAME", right_on="Country", how="left")

fig, ax = plt.subplots(1, 1, figsize=(15,10))
merged.plot(column='GDP (Billions USD)', ax=ax, legend=True,
            legend_kwds={'label': "GDP by Country"},
            cmap='OrRd', missing_kwds={"color": "lightgrey"})

plt.show()
plt.savefig("WorldGDP.png")
```
