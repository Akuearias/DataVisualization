```python
import pandas as pd

df = pd.read_csv("Week10_Example1.csv")
df
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
df_pca = pca.fit_transform(df)
df_pca
```

```python
import seaborn as sns

df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])
sns.jointplot(
    data=df_pca,
    x='PC1', y='PC2',
    kind='scatter',      # æˆ– 'kde'ã€'hex'ã€'hist'
    marginal_kws=dict(fill=True, alpha=0.5)
)
plt.suptitle('PCA Projection with Marginal KDE', y=1.02)
plt.savefig('Exercise12_graph.png')
```
