```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Week4_example.csv')
df.head
```

```python
plt.plot(df['Date'], df['Category_A'], color='red', label='A')
plt.plot(df['Date'], df['Category_B'], color='green', label='B')
plt.plot(df['Date'], df['Category_C'], color='blue', label='C')
plt.plot(df['Date'], df['Category_D'], color='orange', label='D')
```

```python
plt.plot(df['Date'], df['Category_A'], color='red', label='A')
plt.plot(df['Date'], df['Category_B'], color='maroon', label='B')
plt.plot(df['Date'], df['Category_C'], color='magenta', label='C')
plt.plot(df['Date'], df['Category_D'], color='crimson', label='D')
```

```python
import seaborn as sns

df_no_date = df.drop(columns=['Date'])

corr = df_no_date.corr()

plt.figure(figsize=(16, 9))
sns.heatmap(
    corr,
    annot=True,       
    fmt=".2f",         
    cmap="coolwarm",   
    center=0,          
    square=True        
)

plt.show()
```
